Print all player and ranking information of players in the top 10 ranking

```python
%%writefile task2.sqlite3
--.headers on
--.mode columns

select *
from player
join ranking
on ranking.FK_playerid = player.playerid
order by ranking.rank asc
limit 10
```

```python
! sqlite3 hw2tennis.db < task2.sqlite3
```
